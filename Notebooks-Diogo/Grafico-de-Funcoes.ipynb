 # Grafico de Funcao
 Desenhando o grafico da funcao abaixo:

 $y = \frac{x^2}{(x-1)}$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2/(x-1)

N = 4.5
x1 = np.linspace(-N,0.99,1000)
x2 = np.linspace(1.01,2*N,1000)
x = np.hstack((x1,x2))
y = x+1
v = y/y

plt.plot(x1,f(x1),'b')
plt.plot(x2,f(x2),'b')
plt.plot(x,y,'gray')
plt.plot(v,y,'gray')
plt.xlim([-N,2*N])
plt.ylim([-N,2*N])
plt.axhline(color='gray',linewidth=2,linestyle='--')
plt.axvline(color='gray',linewidth=2,linestyle='--')
plt.show()
```
